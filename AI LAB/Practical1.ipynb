```python
""" 
   DFS Water jug problem
"""

from collections import deque

def solve_water(capacity_jug1, capacity_jug2, target):
    queue = deque([(0, 0, [])])
    checkpoint = set()
    
    while queue:
        current_jug1, current_jug2, mark = queue.popleft()
        
        if current_jug1 == target or current_jug2 == target:
            return mark
        
        if (current_jug1, current_jug2) in checkpoint:
            continue
        
        checkpoint.add((current_jug1, current_jug2))
        
        actions = [
            ('fill jug 1', capacity_jug1, current_jug2),
            ('fill jug 2', current_jug1, capacity_jug2),
            ('empty jug 1', 0, current_jug2),
            ('empty jug 2', current_jug1, 0),
            ('pour jug 1 to jug 2', max(current_jug1 - (capacity_jug2 - current_jug2), 0),
             min(current_jug1 + current_jug2, capacity_jug2)),
            ('pour jug 2 to jug 1', min(current_jug1 + current_jug2, capacity_jug1),
             max(current_jug2 - (capacity_jug1 - current_jug1), 0))
        ]
        
        for action, new_jug1, new_jug2 in actions:
            queue.append((new_jug1, new_jug2, mark + [action]))
    
    return None

capacity_jug1 = 5
capacity_jug2 = 6
target = 2

output = solve_water(capacity_jug1, capacity_jug2, target)

if output:
    print("Solution found")
    print("Steps:", len(output))
    print("Path:", output)
else:
    print("Solution not found")
```
